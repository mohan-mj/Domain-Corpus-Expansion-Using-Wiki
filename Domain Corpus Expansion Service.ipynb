```python
from textblob import TextBlob
import pandas as pd
import re
import csv
import wikipedia
```

```python
def load_data(filepath):
    data_list=[]
    df = pd.read_csv(filepath,encoding="utf-8")
    requirements_list = df.Requirement.tolist()
    requirements = []
    for sentence in requirements_list:
        requirements.append(re.sub(r'[^\w\s]',' ',sentence).lower())
    return requirements
```

```python
requirements = load_data(".\dataset\SRS_FR_NFR.csv")
```

```python
requirements
```

```python
noun_phrases_with_duplicates = []
for sentence in requirements:
    blob = TextBlob(sentence)
    noun_phrases_with_duplicates.append(blob.noun_phrases)
```

```python
single_list_noun_phrases=[]
for sublist in noun_phrases_with_duplicates:
    for item in sublist:
        single_list_noun_phrases.append(item)
```

```python
noun_phrases = list(set(single_list_noun_phrases))
```

```python
noun_phrases
```

```python
nouns_with_duplicates = []
for sentence in requirements:
    blob = TextBlob(sentence)
    nouns_with_duplicates.append(blob.tags)
```

```python
print(nouns_with_duplicates)
```

```python
single_list_nouns=[]
for i in nouns_with_duplicates:
    for word,pos in i:
        if (pos == 'NN'):
            single_list_nouns.append(word)
```

```python
nouns = list(set(single_list_nouns))
```

```python
print(nouns)
```

```python
keywords_with_duplicates = noun_phrases + nouns
keywords =  list(set(keywords_with_duplicates))
```

```python
print("Number of NounPhrases: ",len(noun_phrases))
print("Number of Nouns: ",len(nouns))
print("Number of keywords: ",len(keywords))
```

```python
keywords = nouns
```

```python
wikipedia.set_lang('en')
```

```python
def storeResult(wikipage):
    title=wikipage.title
    summary=wikipage.summary.replace('\n',' ')
    appendList= title,summary
    return appendList    
```

```python
fileout=open('wiki_corpus_NN_summary.csv', 'w', encoding="utf8",newline="")
writer = csv.writer(fileout,csv.QUOTE_MINIMAL)
writer.writerow(['Title','Summary'])
count = 0
for data in keywords:
    count = count + 1
    if count % 5 == 0:
        print("fetched summary upto %d keywords out of %d"%(count ,len(keywords)))
    suggestedList =[]
    try:
        wikipage = wikipedia.page(data)
        writer.writerow(storeResult(wikipage))
    except wikipedia.exceptions.DisambiguationError as e:
        suggestedList = e.options[:5]
    except wikipedia.exceptions.PageError:
        suggestedList = wikipedia.search(data)
    for lst in suggestedList[:5]:
        try:
            wikipage = wikipedia.page(lst)
            writer.writerow(storeResult(wikipage))
        except wikipedia.exceptions.DisambiguationError:
            pass
        except wikipedia.exceptions.PageError:
            pass
fileout.close()
```

```python
def getcategories(data):
    categories = []
    suggestedList =[]
    try:
        wikipage = wikipedia.page(data)
        categories.append(wikipage.categories)
    except wikipedia.exceptions.DisambiguationError as e:
        suggestedList = e.options[:5]
    except wikipedia.exceptions.PageError:
        suggestedList = wikipedia.search(data)
    except KeyError:
        pass
    for lst in suggestedList[:5]:
        try:
            wikipage = wikipedia.page(lst)
            categories.append(wikipage.categories)
        except wikipedia.exceptions.DisambiguationError:
            pass
        except wikipedia.exceptions.PageError:
            pass
        except KeyError:
            pass
    return categories
```

```python
categories=[]
count = 0
for keyword in nouns:
    count = count + 1
    if count % 5 == 0:
        print("fetched categories upto %d keywords out of %d"%(count ,len(nouns)))
    categories.extend(getcategories(keyword))
```

```python
single_list_categories=[]
for sublist in categories:
    for item in sublist:
            single_list_categories.append(item)
```

```python
categories = list(set(single_list_categories))
```

```python
print(categories)
```

```python
print("Number of Categories for %d nouns: %d"%(len(nouns),len(categories)))
```

```python
fileout=open('wiki_corpus_NN_categories.csv', 'w', encoding="utf8",newline="")
writer = csv.writer(fileout,csv.QUOTE_MINIMAL)
writer.writerow(["Categories"])
for data in categories:
    writer.writerow([data])
fileout.close()
```
